```python
#importing useful libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/housing.csv")#uploading a csv file as input dataset
data
data.dropna(inplace=True)# removing all the null data
data.info# taking information regarding the data
#now dividing data into test and training data, i.e., X and y
```

```python
#data exploration
from sklearn.model_selection import train_test_split
X = data.drop(['median_house_value'], axis=1)#column therefore axis=1, dependent dataset
y = data['median_house_value']#independent dataset
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2)
```

```python
train_data= X_train.join(y_train) #combining both the training data
train_data
```
